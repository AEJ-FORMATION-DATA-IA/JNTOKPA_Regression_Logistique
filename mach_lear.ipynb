```python
import numpy as np
import pandas as pd
import matplotlib as plt
import sklearn.model_selection as model_selection
from sklearn import linear_model
from sklearn.metrics import confusion_matrix
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
```

```python
#charger le jeu de donnees
data = pd.read_excel('Copie de Coeur.xlsx')
df = data.copy()
```

```python
#verifier s'il y a des doublons
df.duplicated()

#supprimer les doublons
df = df.drop_duplicates()
```

```python
#verifier si doublons (si une ligne se repete)
df.duplicated().value_counts()

#suppriler les doublons
df = df.drop_duplicates()
df
```

```python
#verifier si valeur manquante
df.isna().value_counts()

#Supprimer les données manquantes
df.dropna()
```

```python
#verifier s'il y a des constantes puis (supprimer si oui)
df.nunique()
```

```python
#liste des variables quantitatives
quant_var = df.select_dtypes('int64', 'float32')

#je supprime la variable CŒUR de quant_var
quant_var = quant_var.drop(['CŒUR'], axis= 1)

#liste des variables qualitatives
qual_var = df.select_dtypes('object')
```

```python
#normaliser les variables quantitatives 
def normaliser(datf):
    for col in quant_var:
        datf[col] = datf[col]/datf[col].max()
    return datf

normaliser(df)
```

```python
#recoder les variables qualitatiives

#fonction permettant de recoder une serie
def serie_recoder(serie):
    return serie.astype('category').cat.codes

#fonction permettant de recoder toutes les variables qualitatiives du dataframe
def data_recoder_qual_var(datf):
    for col in qual_var:
        datf[col] = serie_recoder(datf[col])
    return datf

df = data_recoder_qual_var(df)
df
```

```python
#creer un dataframe x qui va contenir de age a pente et y coeur
X = df.iloc[:,:-1]
y= df.CŒUR
```

```python
#train_test_split() fonction on va lui donner en parametre x et y et le taux 0.3
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.3, random_state = 0)
```

```python
print ("X_train: ")
X_train
```

```python
print ("y_train: ")
y_train
```

```python
print("X_test: ")
X_test
```

```python
print("y_test: ")
y_test
```

```python
#instanciation du modele
regLog_mod = linear_model.LogisticRegression(random_state = 0)
```

```python
#entrainer le modele
regLog_mod.fit(X_train, y_train)
```

```python
#score entrainement
regLog_mod.score(X_train, y_train)*100
```

```python
#score test
regLog_mod.score(X_test, y_test)*100
```

```python
#construire la matrice de confusion 
confusion_matrix(y_test, regLog_mod.predict(X_test))
```

```python
#la precision 
precision_score(y_test, regLog_mod.predict(X_test))*100
```

```python
#la sensibilite
recall_score(y_test, regLog_mod.predict(X_test))*100
```
